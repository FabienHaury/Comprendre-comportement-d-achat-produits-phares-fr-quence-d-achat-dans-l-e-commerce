### KPI

```python
# Chiffre d'affaires
requete = """
SELECT SUM(payment_value) AS chiffre_affaires
FROM order_payments_dataset
"""
df = request_sql(requete, conn)
df["chiffre_affaires"] = df["chiffre_affaires"].apply(
    lambda x: f"{x:,.0f}".replace(",", " ")
)
df
```

```python
# Nombre total de commandes
requete = """
SELECT COUNT(DISTINCT order_id) AS nb_commandes
FROM orders_dataset
"""
df = request_sql(requete, conn)
df
```

```python
# Nombre de produits vendus
requete = """
SELECT COUNT(*) AS nb_produits_vendus
FROM order_items_dataset
"""
df = request_sql(requete, conn)
df
```

```python
# Panier moyen
requete = """
SELECT REPLACE(printf('%.2f R$', round(SUM(payment_value) / COUNT(DISTINCT order_id), 2)), ',', ' ') AS panier_moyen
FROM order_payments_dataset
"""
df = request_sql(requete, conn)
df
```

```python
# Trouver le nombre total distinct de clients.
requete = """
SELECT count(DISTINCT customer_unique_id) as "Total Nb client"
FROM customers_dataset
"""
df = request_sql(requete, conn)
df
```

```python
# Lister toutes les villes où il y a des vendeurs différents.
requete = """
SELECT distinct(seller_city)
FROM sellers_dataset
"""
df = request_sql(requete, conn)
df
```

```python
# Trouver la somme totale des paiements réalisés par carte de crédit.
requete = """
SELECT sum(payment_value) as "Total"
FROM order_payments_dataset
where payment_type = "credit_card"
"""
df = request_sql(requete, conn)
df["Total"] = df["Total"].apply(lambda x: f"{x:,.0f}".replace(",", " "))
df
```

```python
# Calculer la moyenne des frais de port de tous les articles commandés.
requete = """
SELECT round(avg(freight_value), 2) as "Frais de port moyen"
FROM order_items_dataset
"""
df = request_sql(requete, conn)
df
```

```python
# Quels est le top 5 des vendeurs ? Avec total de vente et total revenu

requete = """
select seller_id as "Seller", count(order_item_id) as "Total vente", REPLACE(printf('%,d R$', ROUND(SUM(price))), ',', ' ') AS "Total revenue"
from order_items_dataset
group by seller_id
order by "Total vente" desc
limit 5
"""

df = request_sql(requete, conn)
df
```

```python
# Compter le nombre de commandes par état de livraison et ordonner par le plus fréquent.
requete = """ 
select order_status, count(order_status) as "Total commande"
from orders_dataset
group by order_status
order by "Total commande" desc
"""


df = request_sql(requete, conn)
df
```

```python
# Donner la catégorie de produit avec le plus grand nombre d’articles vendus.
requete = """ 
select product_category_name_translation.product_category_name, 
    product_category_name_translation.product_category_name_english, 
    SUM(order_items_dataset.order_item_id) AS "Total vendu"
FROM order_items_dataset
    INNER JOIN products_dataset
        ON order_items_dataset.product_id = products_dataset.product_id
    INNER JOIN product_category_name_translation
        ON products_dataset.product_category_name = product_category_name_translation.product_category_name
group by product_category_name_translation.product_category_name
order by "Total vendu" desc
"""


df = request_sql(requete, conn)
df
```

```python
# Lister les 10 clients ayant dépensé le plus (paiements cumulés).
requete = """ 
select customer_unique_id, 
    sum(payment_value) as "Total depense"
from orders_dataset
    left join order_payments_dataset
        on orders_dataset.order_id = order_payments_dataset.order_id
    left join customers_dataset
        on orders_dataset.customer_id = customers_dataset.customer_id
group by customer_unique_id
order by "Total depense" desc
limit 10
"""


df = request_sql(requete, conn)
df
```

```python
# Obtenir la moyenne du délai (en jours) entre la confirmation de commande et la livraison au client.
requete = """ 
select round(avg(julianday(order_delivered_customer_date) - julianday(order_approved_at))) AS 'Temps moyen de livraison'
from orders_dataset
"""


df = request_sql(requete, conn)
df
```

```python
# Quel est le top 5 des clients ? Avec le nombre d'achat et le montant total et date de premier et dernier achat.

requete = """ 
select customers_dataset.customer_unique_id as "Customer unique ID", 
    count(order_items_dataset.product_id) as "Nb achats", 
    REPLACE(printf('%,d R$', ROUND(SUM(order_items_dataset.price))), ',', ' ') as "Montant total",
    min(orders_dataset.order_purchase_timestamp) as "Date premier achat",
    max(orders_dataset.order_purchase_timestamp) as "Date dernier achat"
from orders_dataset
    left join customers_dataset
        on orders_dataset.customer_id = customers_dataset.customer_id
    left join order_items_dataset
        on orders_dataset.order_id = order_items_dataset.order_id
group by "Customer unique ID"
order by "Nb achats" desc
limit 5
"""


df = request_sql(requete, conn)
df
```

```python
# Pour chaque ville, trouver le nombre total d’acheteurs et le panier moyen.
requete = """ 
with achats_par_ville as (
    select
        customer_city as "Ville d'achat",
        count(distinct customer_unique_id) as "Acheteurs distinct",
        round(avg(payment_value)) as "Panier moyen"
    from customers_dataset
        left join orders_dataset
            on customers_dataset.customer_id = orders_dataset.customer_id
        left join order_payments_dataset
            on orders_dataset.order_id = order_payments_dataset.order_id
    group by customer_city
)
select * 
from achats_par_ville
order by "Acheteurs distinct" desc
"""


df = request_sql(requete, conn)
df
```

```python
# Lister les vendeurs les mieux notés (note = 5).
requete = """ 
with best_vendeurs as (
    select
        seller_id,
        review_score
    from order_reviews_dataset
        left join orders_dataset
            on order_reviews_dataset.order_id = orders_dataset.order_id

         left join order_items_dataset
            on orders_dataset.order_id = order_items_dataset.order_id

)
select * 
from best_vendeurs
where review_score = 5
"""


df = request_sql(requete, conn)
df
```

```python
# Trouver toutes les commandes ayant dépassé la date limite d’estimation de livraison.
requete = """ 
with "date depasse" as (
select order_id, 
    round(julianday(order_delivered_customer_date) - julianday(order_estimated_delivery_date)) as "Depassement"
from orders_dataset
where "Depassement" > 0
)
select *
from "date depasse"
order by "Depassement" desc
"""


df = request_sql(requete, conn)
df
```

```python
# Trouver les vendeurs qui ont le plus de délai (> 30)
requete = """ 
with date_depasse as (
select order_id, 
    round(julianday(order_delivered_customer_date) - julianday(order_estimated_delivery_date)) as "Depassement"
from orders_dataset
where "Depassement" > 0
)

select order_items_dataset.seller_id, round(avg(date_depasse.Depassement)) as "Depassement moyen"
from date_depasse
    inner join order_items_dataset
    on date_depasse.order_id = order_items_dataset.order_id
group by order_items_dataset.seller_id
having "Depassement moyen" > 30
order by "Depassement moyen" desc

"""


df = request_sql(requete, conn)
df
```

```python
# Analyser le taux de réachat par client (combien de clients achètent plus d’une fois).
requete = """ 
with clients_achats as (
  select customer_unique_id, count(order_id) as total_achats
  from customers_dataset
  left join orders_dataset
    on customers_dataset.customer_id = orders_dataset.customer_id
  group by customer_unique_id
)
select 
  (select count(*) from clients_achats where total_achats > 1) * 100.0 /
  (select count(*) from clients_achats where total_achats >= 1) 
  as taux_reachat
"""


df = request_sql(requete, conn)
df
```

```python
# Par produit, analyser la part de livraisons en retard par vendeur.
requete = """ 
select 
    products_dataset.product_category_name,
    seller_id,
    100.0 * SUM(CASE WHEN order_delivered_customer_date > order_estimated_delivery_date THEN 1 ELSE 0 END) / COUNT(*) AS percent_late
from orders_dataset
    inner join order_items_dataset
    on orders_dataset.order_id = order_items_dataset.order_id
    inner join products_dataset
    on order_items_dataset.product_id = products_dataset.product_id
GROUP BY products_dataset.product_category_name, seller_id
having products_dataset.product_category_name <> 'None'
order by percent_late desc
"""


df = request_sql(requete, conn)
df
```

```python
# RFM !
requete = """ 
with recency as (
select 
    customer_unique_id,
    round(julianday("2018-12-31 23:59:59") - julianday(max(order_purchase_timestamp))) as "Recency"
from orders_dataset
    inner join order_payments_dataset
    on orders_dataset.order_id = order_payments_dataset.order_id
    inner join customers_dataset
    on orders_dataset.customer_id = customers_dataset.customer_id
group by customer_unique_id
),

frequency as (
select
    customer_unique_id,
    count(customers_dataset.customer_id) as "Frequency"
from orders_dataset
    inner join order_payments_dataset
    on orders_dataset.order_id = order_payments_dataset.order_id
    inner join customers_dataset
    on orders_dataset.customer_id = customers_dataset.customer_id
group by customer_unique_id
),

monetary as(
select 
    customer_unique_id,
    sum(payment_value) as "Monetary"
from orders_dataset
    inner join order_payments_dataset
    on orders_dataset.order_id = order_payments_dataset.order_id
    inner join customers_dataset
    on orders_dataset.customer_id = customers_dataset.customer_id
group by customer_unique_id
),

rfm as (
select
    recency.customer_unique_id,
    recency.Recency,
    frequency.Frequency,
    monetary.Monetary,
    ntile(5) over(order by recency.Recency desc) as recency_score,
    ntile(5) over(order by frequency.Frequency) as frequency_score,
    ntile(5) over(order by monetary.Monetary) as monetary_score
from recency
    inner join frequency on recency.customer_unique_id = frequency.customer_unique_id
    inner join monetary on recency.customer_unique_id = monetary.customer_unique_id
order by recency.Recency
),

rfm_score as (
select
    customer_unique_id,
    recency_score,
    frequency_score,
    monetary_score,
    recency_score + frequency_score + monetary_score as "RFM_score"
from rfm
),

rfm_segment as (
select 
    customer_unique_id,
    case 
        when RFM_score >= 12 then "Loyal"
        when RFM_score between 8 and 11 then "Potential loyalist"
        when RFM_score between 5 and 7 then "Potential lost"
        else "Lost"
    end as "Segment"
from rfm_score
)

select 
    Segment,
    count(*) as Total
from rfm_segment
group by Segment
"""


df = request_sql(requete, conn)
df
```
